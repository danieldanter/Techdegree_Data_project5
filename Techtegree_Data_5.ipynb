# 2: What is the average number of pets found in a month in the year 2015? Are there months where there is a higher number of animals found?

```python
# change datatime to date type
intakes['datetime'] = pd.to_datetime(intakes['datetime'])

# get the columns year and month
intakes['year'] = intakes['datetime'].apply(lambda x: "%d" % (x.year))
intakes['month'] = intakes['datetime'].apply(lambda x: "%d" % (x.month))

# we only need the year 2015
intakes.loc[intakes['year'] == "2015"]

# get the mean of the value count of the differnt month
month_found = intakes['month'].value_counts().to_frame().mean()
month_found

# final answer:  6682.25  is the average number of pets found in a month in the year 2015
```

# 5: What are the adoption rates for specific breeds?
## Find the top 5 dog breeds in the shelter (based on count) and then find the adoption percentage of each breed.

```python
# first get all Dogs than count frequency with value count
dogs = outcomes.loc[outcomes['animal_type'] == 'Dog']
top_dogs = dogs['breed'].value_counts().to_frame()
top_dogs.head(5)

# final answer: see below
```

```python
# get a list of top 5 breeds and iterate over it -> get value count of each out_comte type nomralize it and get Adoption
L = dogs['breed'].value_counts().index[:5].tolist()
for dog in L:    
    this_dog = dogs.loc[dogs['breed'] == dog]
    diff_pit = this_dog['outcome_type'].value_counts(normalize=True)
    rate = diff_pit['Adoption'] 
    print(dog + " adoptionrate: " + str(rate))
    
 # final answer: see below
```

# 6: What are the adoption rates for different colorings?
## Find the top 5 colorings in the shelter (based on count) and then find the adoption percentage of each color.


```python
# first  count frequency of colors with value count
top_color = outcomes['color'].value_counts().to_frame()
top_color.head(5)

# final answer: see below
```

```python
# get a list of top 5 colors and iterate over it -> get value count of each out_comte type nomralize it and get Adoption
L2 = outcomes['color'].value_counts().index[:5].tolist()
for col in L2:
    this_col = outcomes.loc[outcomes['color'] == col]
    diff_pit = this_col['outcome_type'].value_counts(normalize=True)
    rate = diff_pit['Adoption'] 
    print(col + " adoptionrate: " + str(rate))
    
# final answer: see below
```
